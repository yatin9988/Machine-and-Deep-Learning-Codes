```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def drawImg(img):
    print(img.shape)
    plt.axis("off")
    plt.imshow(img,cmap="gray")
    plt.show()
```

```python
cody_bgr = cv2.imread('cody2.jpg')
cody_bgr = cv2.resize(cody_bgr,(100,100))
cody_rgb = cv2.cvtColor(cody_bgr,cv2.COLOR_BGR2RGB)
cody_gray = cv2.cvtColor(cody_bgr,cv2.COLOR_BGR2GRAY)
```

```python
drawImg(cody_rgb)
```

```python
drawImg(cody_gray)
```

```python
def convolution(img,filter_):
    
    h = img.shape[0]
    w = img.shape[1]
    
    f = filter_.shape[0]
    ## activation map
    new_img = np.zeros((h-f+1,w-f+1))
    
    for row in range(new_img.shape[0]):
        for col in range(new_img.shape[1]):
            for i in range(f):
                for j in range(f):
                    new_img[row][col] += img[row+i][col+j]*filter_[i][j]
            
            if new_img[row][col] > 255:
                new_img[row][col] = 255
                
            if new_img[row][col] < 0:
                new_img[row][col] = 0
                
    return new_img                    
```

```python
blur_filter = np.ones((3,3))/9
output1 = convolution(cody_gray,blur_filter)
```

```python
drawImg(cody_gray)
drawImg(output1)
```

```python
sharp_filter = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
output2 = convolution(cody_gray,sharp_filter)
```

```python
drawImg(cody_gray)
drawImg(output2)
```

```python
edge_filter = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
output3 = convolution(cody_gray,edge_filter)
```

```python
drawImg(cody_gray)
drawImg(output3)
```
